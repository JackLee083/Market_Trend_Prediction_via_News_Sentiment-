```python
def get_content(url):
    dic = {}
    res2 = requests.get(url)
    soup2 = BeautifulSoup(res2.text, 'lxml')
    dic['date'] = soup2.select_one('#MainContent_Contents_lbDate').text
    dic['title'] = soup2.select_one('h1').text
    dic['article'] = soup2.select_one('article').text
    return dic
```

```python
domain = 'https://www.moneydj.com/'
news = []
for url in soup.select('.forumgrid tr a'):
    news.append(get_content(domain + url.get('href')))
```

```python
news[0]
```

```python
for page in range(2, 628):  # 從 1 到 242 的頁面
    url = f'https://www.moneydj.com/kmdj/news/newsreallist.aspx?index1={page}&a=mb090000'
    res = requests.get(url)
    soup = BeautifulSoup(res.text, 'lxml')

    for news_link in soup.select('.forumgrid tr a'):
        full_url = domain + news_link.get('href')
        news.append(get_content(full_url))
```

```python
import pandas as pd
df = pd.DataFrame(news)
df
```
